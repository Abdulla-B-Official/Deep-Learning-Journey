# Problem Statement

### Build a ML Model that predicts the median house price based on 13 socio-economic and property-related features from the Boston Housing dataset.

## 1. Import Necessary Libraries

```python
import numpy as np
from tensorflow.keras.datasets import boston_housing
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
```

## 2. Import Dataset

```python
(X_train, y_train), (X_test, y_test) = boston_housing.load_data()
```

### 3. Data Understanding

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

### 4. Data Preparation

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.std().round(),X_train.mean().round()
```

```python
X_test.std().round(),X_test.mean().round()
```

### 5. Model Building

```python
model = Sequential()

model.add(Dense(64, activation='relu', input_shape=(13,))) #For structured data, use input_shape
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))

# Output layer (IMPORTANT)
model.add(Dense(1,))  # No activation → regression
```

```python
model.compile(optimizer='adam',loss='mse', metrics=['mae'])
```

### 6. Model Training || 7. Model Testing || 8. Model Evaluation

```python
model.fit(X_train, y_train,epochs=100,batch_size=16,validation_split=0.2, verbose=2)
```

### 9. Model Prediction (New Observation)

```python
# Example new data (13 features)
new_data = np.array([[0.02, 18, 2.5, 0, 0.45, 6.5, 65, 4.5, 1, 300, 15, 390, 5]])

# Apply same scaling
new_data = scaler.transform(new_data)

# Predict
prediction = model.predict(new_data)

print("Predicted House Price:", prediction)
```

```python
model.save("boston_regression_model.keras")
```
