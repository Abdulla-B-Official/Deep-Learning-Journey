# Problem Statement
Build a machine learning model to accurately predict whether or not the patients in the dataset have diabetes or not ?

=================================================================================

## 1. Import Necessary libraries

```python
import pandas as pd
import keras
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
from sklearn.metrics import confusion_matrix,precision_score,recall_score,r2_score,accuracy_score,classification_report
```

## 2. Import Data

```python
diabetes_data=pd.read_csv("diabetes.csv")
diabetes_data
```

## 3. Data Understanding

```python
diabetes_data.shape
```

```python
diabetes_data.isna().sum()
```

```python
diabetes_data.dtypes
```

## 4. Data Preprocessing

```python
X=diabetes_data.drop('Outcome',axis = 1)
y=diabetes_data [['Outcome']]
```

## 5. Model Building

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

import warnings
warnings.filterwarnings("ignore")
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(5, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=12,stratify=y)
```

```python
X_train.shape ,y_train.shape
```

```python
X_test.shape , y_test.shape
```

## ALERT : Here's where you need to appy STANDARDIZATION

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## 6. Model Traing | 7. Model Testing | 8. Model Evaluation

```python
model.fit(x=X_train,y=y_train,batch_size=16,epochs=50, validation_split=0.2)
```

## 9. Model Prediction

```python
y_pred=model.predict(X_test)
y_pred=(y_pred>0.60).astype(int)
```

```python
print("accuracy_score   :",accuracy_score(y_test,y_pred))
print("r2_score         :",r2_score(y_test,y_pred))
print("precision_score  :",precision_score(y_test,y_pred))
print("recall_score     :",recall_score(y_test,y_pred))
print("confusion_matrix : \n",confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
